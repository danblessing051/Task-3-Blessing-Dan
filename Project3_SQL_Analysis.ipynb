```python
import pandas as pd
import sqlite3

# Load Excel file
df = pd.read_excel("Dataset_Cleaned.xlsx")

# Create SQL database in memory
conn = sqlite3.connect(':memory:')

# Load dataframe into SQL table
df.to_sql('orders', conn, index=False, if_exists='replace')

print("Database created successfully!")
print("Table name: orders")
print("Total rows:", len(df))
```

```python
query = """
SELECT *
FROM orders
LIMIT 10
"""
result = pd.read_sql_query(query, conn)
print(result)
```

```python
query = """
SELECT COUNT(*) AS Total_Orders
FROM orders
"""
result = pd.read_sql_query(query, conn)
print(result)
```

```python
query = """
SELECT SUM(TotalPrice) AS Total_Revenue
FROM orders
"""
result = pd.read_sql_query(query, conn)
print(result)
```

```python
query = """
SELECT
    AVG(TotalPrice) AS Average_Order_Value,
    MIN(TotalPrice) AS Minimum_Order,
    MAX(TotalPrice) AS Maximum_Order
FROM orders
"""
result = pd.read_sql_query(query, conn)
print(result)
```

```python
query = """
SELECT
    OrderStatus,
    COUNT(*) AS Number_of_Orders
FROM orders
GROUP BY OrderStatus
ORDER BY Number_of_Orders DESC
"""
result = pd.read_sql_query(query, conn)
print(result)
```

```python
query = """
SELECT
    Product,
    COUNT(*) AS Total_Orders,
    SUM(TotalPrice) AS Total_Revenue,
    AVG(TotalPrice) AS Avg_Order_Value
FROM orders
GROUP BY Product
ORDER BY Total_Revenue DESC
"""
result = pd.read_sql_query(query, conn)
print(result)
```

```python
query = """
SELECT
    OrderID,
    Product,
    TotalPrice,
    OrderStatus
FROM orders
WHERE OrderStatus = 'Cancelled'
ORDER BY TotalPrice DESC
LIMIT 10
"""
result = pd.read_sql_query(query, conn)
print(result)
```

```python
query = """
SELECT
    OrderID,
    Product,
    TotalPrice,
    PaymentMethod,
    OrderStatus
FROM orders
WHERE TotalPrice > 2000
ORDER BY TotalPrice DESC
"""
result = pd.read_sql_query(query, conn)
print(result)
```

```python
query = """
SELECT
    ReferralSource,
    COUNT(*) AS Total_Orders,
    SUM(TotalPrice) AS Total_Revenue,
    AVG(TotalPrice) AS Avg_Order_Value
FROM orders
GROUP BY ReferralSource
ORDER BY Total_Orders DESC
"""
result = pd.read_sql_query(query, conn)
print(result)
```

```python
query = """
SELECT
    PaymentMethod,
    COUNT(*) AS Total_Orders,
    SUM(TotalPrice) AS Total_Revenue
FROM orders
GROUP BY PaymentMethod
ORDER BY Total_Orders DESC
"""
result = pd.read_sql_query(query, conn)
print(result)
```

```python
query = """
SELECT
    SUBSTR(Date, 1, 7) AS Month,
    COUNT(*) AS Total_Orders,
    SUM(TotalPrice) AS Monthly_Revenue,
    AVG(TotalPrice) AS Avg_Order_Value
FROM orders
GROUP BY Month
ORDER BY Month ASC
"""
result = pd.read_sql_query(query, conn)
print(result)
```

```python
query = """
SELECT
    Product,
    COUNT(*) AS Cancelled_Orders
FROM orders
WHERE OrderStatus = 'Cancelled'
GROUP BY Product
ORDER BY Cancelled_Orders DESC
"""
result = pd.read_sql_query(query, conn)
print(result)
```

```python
query = """
SELECT
    COUNT(*) AS Total_Orders,
    SUM(TotalPrice) AS Total_Revenue,
    AVG(TotalPrice) AS Avg_Order_Value,
    MIN(TotalPrice) AS Min_Order,
    MAX(TotalPrice) AS Max_Order,
    COUNT(CASE WHEN OrderStatus = 'Cancelled' THEN 1 END) AS Cancelled_Orders,
    COUNT(CASE WHEN OrderStatus = 'Delivered' THEN 1 END) AS Delivered_Orders
FROM orders
"""
result = pd.read_sql_query(query, conn)
print(result)
```

## SQL Commands Used
- SELECT — to choose columns
- FROM — to specify table
- WHERE — to filter rows
- GROUP BY — to group categories
- ORDER BY — to sort results
- COUNT() — to count orders
- SUM() — to total revenue
- AVG() — to average values
- MIN() / MAX() — to find extremes
